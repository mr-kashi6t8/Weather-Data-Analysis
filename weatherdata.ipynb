# Q What is the mean value of each coloumn agains weather condition 

```python
data.head()
```

```python
for col in data.columns:
    if col != 'Weather Condition':  # Assuming 'Weather' is the categorical column for grouping
        data[col] = pd.to_numeric(data[col], errors='coerce')

# Verify the conversion
# print(data.dtypes)
data.groupby("Weather Condition").mean()
```
